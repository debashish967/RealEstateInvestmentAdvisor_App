#### Numeric Columns Sanity Checks

```python
import numpy as np
# Columns to check
numeric_cols = ["BHK", "Size_in_SqFt", "Price_in_Lakhs", "Price_per_SqFt",
                "Year_Built", "Floor_No", "Total_Floors", "Age_of_Property",
                "Nearby_Schools", "Nearby_Hospitals"]

sanity_report = {}

for col in numeric_cols:
    col_data = df[col]
    report = {
        "min": col_data.min(),
        "max": col_data.max(),
        "mean": col_data.mean(),
        "median": col_data.median(),
        "missing": col_data.isna().sum(),
        "negative_values": (col_data < 0).sum()
    }
    sanity_report[col] = report

# Convert report to DataFrame for better readability
sanity_df = pd.DataFrame(sanity_report).T
sanity_df
```

#### Fixing Obvious Numeric Problems

```python
# Example rules (we can adjust thresholds later):
df.loc[df['Size_in_SqFt'] < 100, 'Size_in_SqFt'] = np.nan
df.loc[df['Price_in_Lakhs'] <= 0, 'Price_in_Lakhs'] = np.nan
df.loc[df['Price_per_SqFt'] <= 0, 'Price_per_SqFt'] = np.nan
df.loc[df['BHK'] <= 0, 'BHK'] = np.nan
df.loc[df['Total_Floors'] < df['Floor_No'], ['Floor_No','Total_Floors']] = df[['Total_Floors','Floor_No']]

# Recompute Age_of_Property if Year_Built is reasonable
current_year = 2025
df.loc[(df['Year_Built'] > 1800) & (df['Year_Built'] <= current_year), 'Age_of_Property'] = current_year - df['Year_Built']
```

```python
# --- Step 1: Inspect Year_Built ---
year_unique = sorted(df['Year_Built'].unique())
print("Unique Year_Built values (sorted):", year_unique)

# --- Step 2: Check for numeric anomalies in numeric columns ---
numeric_cols = ['BHK','Size_in_SqFt','Price_in_Lakhs','Price_per_SqFt',
                'Year_Built','Floor_No','Total_Floors','Age_of_Property',
                'Nearby_Schools','Nearby_Hospitals']

numeric_summary = []
for col in numeric_cols:
    s = df[col]
    numeric_summary.append({
        'Column': col,
        'min': s.min(),
        'max': s.max(),
        'mean': s.mean(),
        'median': s.median(),
        'missing': s.isna().sum(),
        'negative_values': (s < 0).sum()
    })

import pandas as pd
numeric_summary_df = pd.DataFrame(numeric_summary)
print("\nNumeric Columns Summary:")
print(numeric_summary_df)

# --- Step 3: Fix obvious Year_Built problems ---
# Assume realistic range: 1950 - 2023
year_min, year_max = 1950, 2023
invalid_years = df[(df['Year_Built'] < year_min) | (df['Year_Built'] > year_max)]
print(f"\nNumber of Year_Built entries outside {year_min}-{year_max}: {len(invalid_years)}")

# Fix by clipping to min/max
df['Year_Built'] = df['Year_Built'].clip(lower=year_min, upper=year_max)

print("\nAfter fixing, Year_Built min/max:", df['Year_Built'].min(), df['Year_Built'].max())
```

#### categorical columns sanity checks

```python
# Categorical columns
cat_cols = ['State','City','Locality','Property_Type','Furnished_Status',
            'Public_Transport_Accessibility','Parking_Space','Security',
            'Amenities','Facing','Owner_Type','Availability_Status']

cat_summary = []
for col in cat_cols:
    s = df[col]
    cat_summary.append({
        'Column': col,
        'Unique_Values': s.nunique(),
        'Missing': s.isna().sum(),
        'Top_3_Uniques': s.value_counts().head(3).to_dict()
    })

cat_summary_df = pd.DataFrame(cat_summary)
print("Categorical Columns Summary:")
print(cat_summary_df)
```

#### Fix Price_per_SqFt = 0 by recalculating it from Price_in_Lakhs and Size_in_SqFt.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# -----------------------------
# Fix Price_per_SqFt zeros
# -----------------------------
zero_ppsqft_count = (df['Price_per_SqFt'] == 0).sum()
print(f"Price_per_SqFt = 0 entries: {zero_ppsqft_count}")

# Recalculate: Price_per_SqFt = (Price_in_Lakhs * 100000) / Size_in_SqFt
df.loc[df['Price_per_SqFt'] == 0, 'Price_per_SqFt'] = (
    df.loc[df['Price_per_SqFt'] == 0, 'Price_in_Lakhs'] * 100000 / 
    df.loc[df['Price_per_SqFt'] == 0, 'Size_in_SqFt']
)

# -----------------------------
# Encode categorical columns
# -----------------------------
# Columns for label encoding
label_cols = ['Furnished_Status', 'Public_Transport_Accessibility', 
              'Parking_Space', 'Security', 'Facing', 'Owner_Type', 'Availability_Status']

le_dict = {}
for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    le_dict[col] = le  # store encoder for future inverse transform if needed

# Columns for one-hot encoding
onehot_cols = ['State', 'City', 'Property_Type']
df = pd.get_dummies(df, columns=onehot_cols, drop_first=True)

# -----------------------------
# Optional: drop Locality (too many unique values) for ML
# -----------------------------
df_ml = df.drop(columns=['Locality', 'ID', 'Amenities'])  # Keep amenities for later multi-hot encoding if needed

print("Data cleaning & encoding done. Ready for ML.")
print(f"Shape of cleaned dataset: {df_ml.shape}")

df_ml.head()
```

###### Since Amenities is a multi-label column (comma-separated like Pool, Gym, Garden), we can create multi-hot encoded columns: one column per unique amenity with 1/0.

```python
# -----------------------------
# Multi-hot encode Amenities
# -----------------------------
# Split comma-separated amenities and get unique set
all_amenities = set()
df['Amenities'].dropna().map(lambda x: [a.strip() for a in x.split(',')]).map(all_amenities.update)

print(f"Total unique amenities: {len(all_amenities)}")
print("Some sample amenities:", list(all_amenities)[:10])

# Create multi-hot columns
for amenity in all_amenities:
    df[f"Amenity_{amenity.replace(' ', '_')}"] = df['Amenities'].apply(lambda x: 1 if amenity in x else 0)

# Drop original Amenities column
df.drop('Amenities', axis=1, inplace=True)

print(f"Shape after Amenities encoding: {df.shape}")
```

✅ Use Median (and here’s the short reason)
Use median for Price_per_SqFt because:

Real estate price-per-sqft is highly skewed
(a few luxury houses have extremely high price-per-sqft)

Median is robust to outliers, mean is not

Mean would artificially inflate values for normal properties

Median gives a stable central tendency that fits real estate well

✔ Professional real-estate ML models generally impute median for all skewed price-like columns.

#### ✅ Code to Replace NaN in Price_per_SqFt (Median Imputation)

```python
# Median Imputation for Price_per_SqFt
median_ppsqft = df['Price_per_SqFt'].median()
df['Price_per_SqFt'] = df['Price_per_SqFt'].fillna(median_ppsqft)

# Verify
print("Missing values after fix:", df['Price_per_SqFt'].isna().sum())
```

#### Train ML Model: Gradient Boosting Regressor (Price Prediction)

#### Train ML Model — HistGradientBoostingRegressor (Fast Version)

```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Model
hgb = HistGradientBoostingRegressor(
    learning_rate=0.05,
    max_depth=6,
    max_iter=500,
    random_state=42
)

# Train
hgb.fit(X_train, y_train)

# Predict
y_pred = hgb.predict(X_test)

# Evaluate
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))  # <-- FIXED
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}, R2: {r2:.4f}")
```
